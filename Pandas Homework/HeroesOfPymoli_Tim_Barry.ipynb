## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# drawing data from purchase 
#dropping duplicates

DropDup = purchase_data.drop_duplicates(subset='SN', keep="first")
TotalGen = DropDup["Gender"].count()
MaleGen = DropDup["Gender"].value_counts()['Male']
FemGen = DropDup["Gender"].value_counts()['Female']
NoGen = TotalGen - MaleGen - FemGen

#Create the percentages

MalePer = (MaleGen / TotalGen) * 100
FemPer = (FemGen / TotalGen) * 100
NonPer = (NoGen / TotalGen) * 100

# Create the gender analysis tool as a dataframe

GenderDem = pd.DataFrame({"": ['Male', 'Female', 'Other'],
                            "Percentage of Players": [MalePer, FemPer, NonPer],
                            "Total Count": [MaleGen, FemGen, NoGen]})


# format the dataframe output

GenderDem["Percentage of Players"] = GenderDem["Percentage of Players"].map("{:.2f}%".format)
#remove number column

GenderDem = GenderDem.set_index('')

GenderDem
```
